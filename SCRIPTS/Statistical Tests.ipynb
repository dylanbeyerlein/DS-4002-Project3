```python
#Import necessary libraries
import pandas as pd
from scipy.stats import wilcoxon
from scipy.stats import friedmanchisquare
```

```python
#Upload dataset from the PyTorch models
df = pd.read_csv("/Users/harryshield/Documents/Data Science/DS4002/Project 3/Model_Comparison.csv")
df.head()
```

```python
#Classic Friedman documentation for the evaluation of recall for all 3 models
df_metric = df[df["Evaluation"] == "Recall"].reset_index(drop=True)
df_metric = df_metric[["Evaluation", "Fold", "ResNet18", "EfficientNetB0", "DenseNet121"]]

ResNet18 = df_metric["ResNet18"].values
DenseNet121 = df_metric["DenseNet121"].values
EfficientNetB0 = df_metric["EfficientNetB0"].values


stat, p = friedmanchisquare(ResNet18, DenseNet121, EfficientNetB0)

#The statistic and p value tells us the confidence of the significance
print("Friedman statistic:", stat)
print("p-value:", p)
```

```python
#Friedman for F1
df_metric = df[df["Evaluation"] == "F1"].reset_index(drop=True)
df_metric = df_metric[["Evaluation", "Fold", "ResNet18", "EfficientNetB0", "DenseNet121"]]

ResNet18 = df_metric["ResNet18"].values
DenseNet121 = df_metric["DenseNet121"].values
EfficientNetB0 = df_metric["EfficientNetB0"].values


stat, p = friedmanchisquare(ResNet18, DenseNet121, EfficientNetB0)

print("Friedman statistic:", stat)
print("p-value:", p)
```

```python
# ResNet18 vs. DenseNet121 F1 Wilcoxon pairwise
df_balance = df[df["Evaluation"] == "F1"]
df_balance = df_balance[["Evaluation", "Fold", "ResNet18", "DenseNet121"]]
df_balance = df_balance.reset_index(drop=True)
df_balance.head()

stat, p = wilcoxon(df_balance["ResNet18"], df_balance["DenseNet121"])

#Prints the statistic which tells the strength of the two models, and the confidence of the significance
print("Wilcoxon statistic:", stat)
print("p-value:", p)
```

```python
# ResNet18 vs. EfficientNetB0 F1 Wilcoxon pairwise
df_balance_2 = df[df["Evaluation"] == "F1"]
df_balance_2 = df_balance_2[["Evaluation", "Fold", "ResNet18", "EfficientNetB0"]]
df_balance_2 = df_balance_2.reset_index(drop=True)
df_balance_2.head()

stat, p = wilcoxon(df_balance_2["ResNet18"], df_balance_2["EfficientNetB0"])

print("Wilcoxon statistic:", stat)
print("p-value:", p)
```

```python
# EfficientNetB0 vs. DenseNet121 F1 Wilcoxon pairwise
df_balance_3 = df[df["Evaluation"] == "F1"]
df_balance_3 = df_balance_3[["Evaluation", "Fold", "EfficientNetB0", "DenseNet121"]]
df_balance_3 = df_balance_3.reset_index(drop=True)
df_balance_3.head()

stat, p = wilcoxon(df_balance_3["EfficientNetB0"], df_balance_3["DenseNet121"])

print("Wilcoxon statistic:", stat)
print("p-value:", p)
```
