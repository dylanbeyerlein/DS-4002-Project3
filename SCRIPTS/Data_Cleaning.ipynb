```python
#Import necessary library for data upload and cleaning
import pandas as pd
```

```python
df = pd.read_csv("/Users/harryshield/Documents/Data Science/DS4002/Project 3/ISIC 2020 Metadata.csv")

df.head()

print("Before:", len(df))
#Drop any rows that have blank values
df = df.dropna()

print("After:", len(df))

df = df.reset_index()

df.head()

```

```python
#Sort out the pertinent columns 
df_clean = df[['image_name', 'patient_id','sex', 'anatom_site_general_challenge', 'age_approx', 'target']]

benign = df[df['target'] == 0]
malignant = df[df['target'] == 1]

benign_sample = benign.sample(n=4425, random_state=42)

df_slim = pd.concat([malignant, benign_sample], axis=0)
df_slim = df_slim.sample(frac=1, random_state=42).reset_index(drop=True)

print(df_slim["target"].value_counts())
print("Total rows:", len(df_slim))
```

```python
df_slim.to_csv("/Users/harryshield/Desktop/cleaned_data.csv", index=False)
```
